```python
import pandas as pd
```

# **Datos de Aerosoles**

```python
df_aerosoles = pd.read_csv('data/All_Sites_Times_Daily_Averages_AOD20.csv')
df_aerosoles.tail()
```

```python
columnas_aerosoles = df_aerosoles.columns
print(len(columnas_aerosoles))
print(columnas_aerosoles)
```

## **`Sub-categorias del dataset`**

#### **`Metadatos de las estaciones AERONET`**

```python
metadatos = df_aerosoles.filter(like='AERONET')
print(len(list(metadatos.columns)))
metadatos.head()
```

#### **`Informacion temporal`**

```python
temp = df_aerosoles.filter(like='Time').copy()
temp['Date(dd:mm:yyyy)'] = pd.to_datetime(df_aerosoles['Date(dd:mm:yyyy)'], format='%d:%m:%Y')
temp['Time(hh:mm:ss)'] = pd.to_datetime(df_aerosoles['Time(hh:mm:ss)'], format='%H:%M:%S').dt.time
temp['Day_of_Year'] = df_aerosoles['Day_of_Year']

print(len(list(temp.columns)))
temp.head()
```

```python
df_aerosoles['Time(hh:mm:ss)'].value_counts()
```

```python
valores = temp['Time(hh:mm:ss)'].value_counts()
print(valores)
```

#### **`Informacion espacial`**

```python
esp = df_aerosoles[['Site_Latitude(Degrees)', 'Site_Longitude(Degrees)', 'Site_Elevation(m)']]
print(len(list(esp.columns)))
esp.head()
```

#### **`Conteo de Mediciones`**

```python
conteo = df_aerosoles.filter(like='N')
conteo = conteo.drop(columns=list(metadatos.columns))

conteo.head()
```

```python
columnas_conteo = conteo.columns
print(len(columnas_conteo))
print(columnas_conteo)
```

#### **`Mediciones de AOD (Aerosol Optical Depth) - espesor ptico de aerosoles en diferentes longitudes`**

```python
AOD = df_aerosoles.filter(like='AOD')
AOD = AOD.drop(columns=list(conteo.filter(like='N[AOD').columns))

AOD.head()
```

```python
columnas_AOD = AOD.columns
print(columnas_AOD)
```

```python
# Longitudes de onda qe nos sirven
funcionales = AOD[['AOD_340nm', 'AOD_380nm', 'AOD_400nm', 'AOD_412nm', 'AOD_440nm', 'AOD_443nm', 'AOD_490nm', 'AOD_500nm', 'AOD_510nm', 'AOD_532nm', 'AOD_551nm', 'AOD_555nm', 'AOD_560nm', 'AOD_620nm', 'AOD_667nm', 'AOD_675nm', 'AOD_779nm']]

funcionales = funcionales.replace(-999, pd.NA)
```
